```python
import pandas as pd
import numpy as np
```

```python
prop = pd.read_excel("./KauaiCountyHI-20211202-PropertyMaster.xlsx")
owner = pd.read_excel("./KauaiCountyHI-20211202-OwnerMaster.xlsx")
```

```python
prop = prop.drop(columns="OwnerName")
merged = prop.merge(owner, how="outer", on="ParcelId")
```

```python
city = []
state = []
zip_code = []

for s in merged["OwnerCityStZip"]:
    if not isinstance(s, str) and np.isnan(s):
        city.append(s)
        state.append(s)
        zip_code.append(s)
    else:
        words = s.split(" ")
        zip_code.append(words[-1])
        state.append(words[-2])
        city.append(" ".join(s.split(" ")[:-2]))

merged = merged.drop(columns="OwnerCityStZip")
merged["OwnerCity"] = city
merged["OwnerState"] = state
merged["OwnerZip"] = zip_code
```

```python
merged.to_csv("KauaiCountyHI-20211202-PropertyOwnerMaster.csv")
```
