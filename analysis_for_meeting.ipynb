```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
huliaina = gpd.read_file("./huliaina (2).zip")
```

```python
years = sorted([
    '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2005', '2000', '1995', '1990',
    '1985', '1980', '1975', '1970', '1965', '1960', '1955', '1950', '1945', '1940', '1928',
    '1927', '1926', '1925', '1923', '1912', '1907', '1873', '1851',
])
```

```python
old_props = huliaina[huliaina["1950"] != 0]
```

```python
huliaina[huliaina[""]]
```

```python
%matplotlib notebook
mean_value = []
for year in years[10:]:
    mean_value.append(old_props[year].mean())
plt.plot(years[10:], np.array(mean_value)/1e6)
plt.xticks(rotation=45)
plt.ylabel("mean property value (millions)")
plt.title(f"Values of {len(old_props)} longtime properties, Anini to Kalihiwai (Dominique's data)")
plt.savefig("./figures for meeting/huliaina_longtime_prop_value.png")

# needs to be adjusted for inflation
```

```python
# max property value
%matplotlib notebook
max_value = []
for year in years:
    max_value.append(huliaina[year].max())
plt.plot(years, np.array(max_value)/1e6)
plt.xticks(rotation=45)
plt.ylabel("max property value (millions)")
plt.title(f"Max values of properties, Anini to Kalihiwai (Dominique's data)")
#plt.savefig("./figures for meeting/huliaina_longtime_prop_value.png")

# needs to be adjusted for inflation?
```

```python
corelogic = pd.read_csv("corelogic_with_tmk.csv")
tmk = gpd.read_file("./Parcels_-_Kauai (1).zip")
tmk = tmk[tmk.zone >= 4]
```

```python
df = tmk[["tmk", "geometry"]].merge(corelogic, how="left", on="tmk")
```

```python
list(df.columns)
```

```python
fig, ax = plt.subplots()
df.plot(ax=ax, column="MARKET_TOTAL_VALUE", legend=True,
        legend_kwds={'label': "Market Value (USD)",
                        'orientation': "horizontal"},
       missing_kwds={'color': 'lightgrey'},
       cmap="Reds")
ax.set_xlim([-159.6, -159.28])
ax.set_ylim([22.125, 22.24])
ax.set_xticks([])
ax.set_yticks([])
plt.title("2019 Market Value, Halele'a + Ko'olau")

plt.savefig("./figures for meeting/2019_mkt_val_NS.png")
```

```python
fig, ax = plt.subplots()
df.plot(ax=ax, column="TAX_AMOUNT", legend=True,
        legend_kwds={'label': "Property Tax (USD)",
                        'orientation': "horizontal"},
       missing_kwds={'color': 'lightgrey'},
       cmap="Reds")
ax.set_xlim([-159.6, -159.28])
ax.set_ylim([22.125, 22.24])
ax.set_xticks([])
ax.set_yticks([])
plt.title("2019 Property Tax, Halele'a + Ko'olau")

plt.savefig("./figures for meeting/2019_tax_amount_NS.png")
```

```python
fig, ax = plt.subplots()
df["value_per_acre"] = df["MARKET_TOTAL_VALUE"]/df["ACRES"]
df["value_per_acre"][df["value_per_acre"] > 1e7] = np.nan
df.plot(ax=ax, column="value_per_acre", legend=True,
        legend_kwds={'label': "Property Value per Acre (USD/acre)",
                        'orientation': "horizontal"},
       missing_kwds={'color': 'lightgrey'},
       cmap="Reds")
ax.set_xlim([-159.6, -159.28])
ax.set_ylim([22.125, 22.24])
ax.set_xticks([])
ax.set_yticks([])
plt.title("2019 Property Value per Acre, Halele'a + Ko'olau")

plt.savefig("./figures for meeting/2019_value_per_acre_NS.png")
```
